```
import sys
from PIL import Image
import glob
import os
import os.path
#from tqdm.notebook import tqdm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from tqdm import tqdm_notebook
```

```
#rgb_dir = "/content/drive/My Drive/Deep Learning/Project 2/FLIR/FLIR_ADAS_1_3/train/RGB/"
rgb_dir = "/content/RGB/"
print( len(glob.glob(rgb_dir+'*')) )
print( len([name for name in os.listdir(rgb_dir) if os.path.isfile(os.path.join(rgb_dir, name))]) )
```

```
#code for grayscale
counter = 0
final_height_width = 256
for filename in tqdm_notebook(glob.glob(rgb_dir+'/*.jpg')):
    counter+=1
    #print(str(counter) + "  " + filename)
    splitFile = filename.split('.')
    thermalFile = splitFile[0].replace('RGB','thermal_8_bit') + '.jpeg'
    #print(thermalFile)
    if(os.path.exists(thermalFile)):
        #filename = 'E:/Study/Sem2/Independent Study/FLIR/FLIR_ADAS_1_3/train/RGB/FLIR_00008.jpg'
        #filename_thermal = 'E:/Study/Sem2/Independent Study/FLIR/FLIR_ADAS_1_3/train/thermal_8_bit/FLIR_00008.jpeg'
        #print(filename)
        #print(filename_thermal)
        #resize color image
        rgb_img = cv2.imread(filename)
        
        w, h, c = rgb_img.shape
        #print("width : " , w ," Height : ", h)

        rgb_img = rgb_img[0+169:w-176,0+192:h-40]
        w, h, c = rgb_img.shape
        #print("width : " , w ," Height : ", h)
        
        
        rgb_img = cv2.resize(rgb_img,(640,512),interpolation = cv2.INTER_AREA)
        w, h, c = rgb_img.shape
        rgb_img = rgb_img[:,0+64:h-64]
        rgb_img = cv2.resize(rgb_img,(final_height_width,final_height_width),interpolation = cv2.INTER_AREA)
        
        rgb_file = splitFile[0].replace('RGB','resize/RGB') + '.jpg'
        #rgb_file = filename.replace('RGB','resize/RGB') + '.jpg'
        cv2.imwrite(rgb_file,rgb_img)
        
        
        thermal_img = cv2.imread(thermalFile)
        w, h, c = thermal_img.shape
        thermal_img = thermal_img[:,0+64:h-64]
        thermal_img = cv2.resize(thermal_img,(final_height_width,final_height_width),interpolation = cv2.INTER_AREA)
        
        #thermal_file = filename_thermal.replace('RGB','resize/thermal_8_bit') + '.jpg'
        thermal_file = thermalFile.replace('thermal_8_bit','resize/thermal_8_bit') + '.jpg'
        thermal_file = thermal_file.replace('.jpeg.jpg', '.jpg')
        cv2.imwrite(thermal_file,thermal_img)
        
        #thermal_img.show(command='fim')
        
        #print(counter)
```
